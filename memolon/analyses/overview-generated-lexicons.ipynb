# Overview of TargetPred Lexicons

```python
import numpy as np
import pandas as pd
import json

from memolon.src import utils, constants
```

Read in silver evaluation results:

```python
silver = pd.read_csv(constants.SILVER_EVALUATION_RESULTS, index_col=0)
silver.head()
```

Add information about size of lexicons and full language name:

```python
silver['No. Words'] = 0
```

```python
for iso in silver.index:
    silver.loc[iso, 'No. Words'] = len(utils.get_TargetPred(iso=iso, split='full'))
```

```python
silver
```

```python
langs = utils.language_table
langs
```

```python
silver['language full name'] = ""
```

```python
silver.head()
```

```python
for iso in langs.index:
    full = langs.loc[iso, 'fastText_fullname']
    silver.loc[iso, 'language full name'] = full
     
```

```python
silver.head()
```

Sort by mean silver evaluation result over emotions and format DataFrame:

```python
silver['ISO'] = silver.index
df = silver[['ISO','language full name', 'No. Words', 'valence', 'arousal', 'dominance',
             'joy', 'anger', 'sadness', 'fear', 'disgust']]
df['Mean'] = df[constants.emotions].mean(axis=1)
rn = {'valence': 'Val',
      'arousal': 'Aro', 
      'dominance': 'Dom',
      'joy': 'Joy',
      'anger': 'Ang', 
      'sadness': 'Sad',
      'fear': 'Fea',
      'disgust': 'Dis',
      'language full name': 'Full Name'}

df = df.rename(columns=rn)

df = df.sort_values('Mean', ascending=False)
df = df.round(2)

df['ISO'] = df.index
df.index = range(1, len(df)+1)
df.index.rename('No.', inplace='True')
df.head()

df.to_csv(constants.GENERATED_LEXICA_OVERVIEW)
```
